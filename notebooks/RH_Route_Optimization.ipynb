#Create a clean version of the address.

```python
import re

def clean_address(addr):
    if pd.isna(addr):
        return addr

    addr = str(addr)

    # Remove the content within the parentheses.
    addr = re.sub(r"\(.*?\)", "", addr)

    # Remove the content following "Unit / Apt / Suite".
    addr = re.sub(r"\b(Unit|Apt|Apartment|Suite)\b.*?,", ",", addr, flags=re.IGNORECASE)

    # Remove extra spaces.
    addr = re.sub(r"\s+", " ", addr).strip()

    return addr

customer_orders["Clean_Address"] = customer_orders["Customer Address"].apply(clean_address)

display(customer_orders[["Customer ID", "Customer Address", "Clean_Address"]])
```

#Import OR-Tools

```python
from ortools.constraint_solver import pywrapcp, routing_enums_pb2
```

#Define OR-Tools Routing Functions

```python
def build_distance_matrix(day_df):
    locations = [(warehouse_lat, warehouse_lon)] + list(zip(day_df["Lat"], day_df["Lon"]))

    matrix = []
    for i in range(len(locations)):
        row = []
        for j in range(len(locations)):
            dist = approx_miles(
                locations[i][0], locations[i][1],
                locations[j][0], locations[j][1]
            )
            row.append(int(dist * 1000))  # 转成整数
        matrix.append(row)
    return matrix
```

#Define the solution function.

```python
def solve_daily_route(day_df):
    distance_matrix = build_distance_matrix(day_df)

    manager = pywrapcp.RoutingIndexManager(len(distance_matrix), 1, 0)
    routing = pywrapcp.RoutingModel(manager)

    def distance_callback(from_index, to_index):
        from_node = manager.IndexToNode(from_index)
        to_node = manager.IndexToNode(to_index)
        return distance_matrix[from_node][to_node]

    transit_callback_index = routing.RegisterTransitCallback(distance_callback)
    routing.SetArcCostEvaluatorOfAllVehicles(transit_callback_index)

    search_parameters = pywrapcp.DefaultRoutingSearchParameters()
    search_parameters.first_solution_strategy = routing_enums_pb2.FirstSolutionStrategy.PATH_CHEAPEST_ARC

    solution = routing.SolveWithParameters(search_parameters)

    if not solution:
        return None

    route_nodes = []
    index = routing.Start(0)

    while not routing.IsEnd(index):
        node = manager.IndexToNode(index)
        route_nodes.append(node)
        index = solution.Value(routing.NextVar(index))

    route_nodes.append(manager.IndexToNode(index))
    return route_nodes
```

#The First Edition of "Divisional Logic"

```python
def calculate_day_metrics(day_customer_ids, customer_df, warehouse_lat, warehouse_lon, avg_speed_mph=30):
    """
Given a list of customer IDs for a specific day:
1. Use OR-Tools to determine the optimal sequence.
2. Calculate the actual total mileage, total travel time, total service time, and total working hours.
    """
    if len(day_customer_ids) == 0:
        return {
            "ordered_customers": [],
            "total_drive_miles": 0,
            "total_drive_minutes": 0,
            "total_service_minutes": 0,
            "total_work_minutes": 23
        }

    day_df = customer_df[customer_df["Customer ID"].isin(day_customer_ids)].copy().reset_index(drop=True)

    route_nodes = solve_daily_route(day_df)
    if route_nodes is None:
        return None

    ordered_customers = [
        day_df.iloc[node - 1]["Customer ID"]
        for node in route_nodes
        if node != 0
    ]

    ordered_df = customer_df[customer_df["Customer ID"].isin(ordered_customers)].copy()
    ordered_df["route_order"] = ordered_df["Customer ID"].map(
        {cust_id: i for i, cust_id in enumerate(ordered_customers)}
    )
    ordered_df = ordered_df.sort_values("route_order").reset_index(drop=True)

    total_drive_miles = 0
    total_drive_minutes = 0
    total_service_minutes = 0

    prev_lat, prev_lon = warehouse_lat, warehouse_lon

    for _, row in ordered_df.iterrows():
        leg_miles = approx_miles(prev_lat, prev_lon, row["Lat"], row["Lon"])
        leg_minutes = leg_miles / avg_speed_mph * 60

        total_drive_miles += leg_miles
        total_drive_minutes += leg_minutes

        # Service Time per Stop = Total Loading/Unloading Time + Fixed Stop Time
        stop_fixed = 14
        service_minutes = row["Total_Service_Time"] + stop_fixed
        total_service_minutes += service_minutes

        prev_lat, prev_lon = row["Lat"], row["Lon"]

    # Final stop: back to the warehouse.
    return_miles = approx_miles(prev_lat, prev_lon, warehouse_lat, warehouse_lon)
    return_minutes = return_miles / avg_speed_mph * 60

    total_drive_miles += return_miles
    total_drive_minutes += return_minutes

    total_work_minutes = 23 + total_drive_minutes + total_service_minutes

    return {
        "ordered_customers": ordered_customers,
        "total_drive_miles": total_drive_miles,
        "total_drive_minutes": total_drive_minutes,
        "total_service_minutes": total_service_minutes,
        "total_work_minutes": total_work_minutes
    }
```

```python
MAX_VOLUME = 1700
MAX_WEIGHT = 7000
MAX_MINUTES = 540
AVG_SPEED_MPH = 30

days = []
remaining_ids = unassigned["Customer ID"].tolist()

while remaining_ids:
    # First, select a seed customer (the largest one remaining).
    seed_id = remaining_ids[0]
    current_day = [seed_id]
    remaining_ids.remove(seed_id)

    improved = True

    while improved and remaining_ids:
        improved = False
        best_candidate = None
        best_metrics = None
        best_score = -1

        for cust_id in remaining_ids:
            candidate_day = current_day + [cust_id]

            candidate_df = regular_customers[
                regular_customers["Customer ID"].isin(candidate_day)
            ].copy()

            total_volume = candidate_df["Total_Volume"].sum()
            total_weight = candidate_df["Total_Weight"].sum()

            # First, check the capacity.
            if total_volume > MAX_VOLUME or total_weight > MAX_WEIGHT:
                continue

            # Re-verify actual working hours (including travel time between clients + return to depot).
            metrics = calculate_day_metrics(
                candidate_day,
                regular_customers,
                warehouse_lat,
                warehouse_lon,
                AVG_SPEED_MPH
            )

            if metrics is None:
                continue

            if metrics["total_work_minutes"] <= MAX_MINUTES:
                # Scoring: Prioritize making the day "fuller"
                # Think of it this way: the closer to 540, the better.
                score = metrics["total_work_minutes"]

                if score > best_score:
                    best_score = score
                    best_candidate = cust_id
                    best_metrics = metrics

        if best_candidate is not None:
            current_day.append(best_candidate)
            remaining_ids.remove(best_candidate)
            improved = True

    days.append(current_day)

for i, d in enumerate(days, 1):
    print(f"Day {i}: {d}")
```

```python
def pack_trips_into_workdays(trips, customer_df, warehouse_lat, warehouse_lon, max_minutes=540, avg_speed_mph=30):
    """
    Re-pack existing trips into true workdays.
    Logic:
    - Treat each element in `trips` as one complete truck trip:
        Warehouse -> customers -> Warehouse
    - Multiple trips can be assigned to the same workday
      as long as cumulative work time <= max_minutes.

    Inputs
    ------
    trips : list[list[str]]
        Each inner list is a trip made of customer IDs.
    customer_df : pd.DataFrame
        Must contain Customer ID, Total_Volume, Total_Weight, Total_Service_Time, Lat, Lon.
    warehouse_lat, warehouse_lon : float
        Depot coordinates.
    max_minutes : int
        Max total work minutes per day, default 540.
    avg_speed_mph : float
        Travel speed assumption.

    Returns
    -------
    workdays : list[list[list[str]]]
        Structure:
        [
          [trip1, trip2],   # Workday 1 has 2 trips
          [trip3],          # Workday 2 has 1 trip
          ...
        ]
    """

    def trip_metrics(customer_ids):
        """
        Compute full trip metrics:
        Warehouse -> customers -> Warehouse
        using OR-Tools route inside the trip.
        """
        if len(customer_ids) == 0:
            return {
                "ordered_customers": [],
                "total_drive_miles": 0,
                "total_drive_minutes": 0,
                "total_service_minutes": 0,
                "total_work_minutes": 0,
                "total_volume": 0,
                "total_weight": 0
            }

        trip_df = customer_df[customer_df["Customer ID"].isin(customer_ids)].copy()

        total_volume = trip_df["Total_Volume"].sum()
        total_weight = trip_df["Total_Weight"].sum()

        # Single trip itself still cannot violate truck capacity
        if total_volume > MAX_VOLUME or total_weight > MAX_WEIGHT:
            return None

        # Use existing OR-Tools routine to optimize the stop sequence
        day_df = trip_df.reset_index(drop=True)
        route_nodes = solve_daily_route(day_df)

        if route_nodes is None:
            return None

        ordered_customers = [
            day_df.iloc[node - 1]["Customer ID"]
            for node in route_nodes
            if node != 0
        ]

        ordered_df = customer_df[customer_df["Customer ID"].isin(ordered_customers)].copy()
        ordered_df["route_order"] = ordered_df["Customer ID"].map(
            {cust_id: i for i, cust_id in enumerate(ordered_customers)}
        )
        ordered_df = ordered_df.sort_values("route_order").reset_index(drop=True)

        total_drive_miles = 0
        total_drive_minutes = 0
        total_service_minutes = 0

        prev_lat, prev_lon = warehouse_lat, warehouse_lon

        for _, row in ordered_df.iterrows():
            leg_miles = approx_miles(prev_lat, prev_lon, row["Lat"], row["Lon"])
            leg_minutes = leg_miles / avg_speed_mph * 60

            total_drive_miles += leg_miles
            total_drive_minutes += leg_minutes

            stop_fixed = 14
            service_minutes = row["Total_Service_Time"] + stop_fixed
            total_service_minutes += service_minutes

            prev_lat, prev_lon = row["Lat"], row["Lon"]

        # return to warehouse
        return_miles = approx_miles(prev_lat, prev_lon, warehouse_lat, warehouse_lon)
        return_minutes = return_miles / avg_speed_mph * 60

        total_drive_miles += return_miles
        total_drive_minutes += return_minutes

        # include 23 min startup for each trip
        total_work_minutes = 23 + total_drive_minutes + total_service_minutes

        return {
            "ordered_customers": ordered_customers,
            "total_drive_miles": total_drive_miles,
            "total_drive_minutes": total_drive_minutes,
            "total_service_minutes": total_service_minutes,
            "total_work_minutes": total_work_minutes,
            "total_volume": total_volume,
            "total_weight": total_weight
        }

    # First compute metrics for every existing trip
    trip_pool = []
    for idx, trip in enumerate(trips, start=1):
        metrics = trip_metrics(trip)
        if metrics is None:
            print(f"Trip {idx} infeasible and skipped: {trip}")
            continue

        trip_pool.append({
            "trip_id": idx,
            "customers": metrics["ordered_customers"],
            "trip_work_minutes": metrics["total_work_minutes"],
            "trip_drive_miles": metrics["total_drive_miles"],
            "trip_drive_minutes": metrics["total_drive_minutes"],
            "trip_service_minutes": metrics["total_service_minutes"],
            "trip_volume": metrics["total_volume"],
            "trip_weight": metrics["total_weight"]
        })

    # Sort trips by descending trip duration so large trips are packed first
    trip_pool = sorted(trip_pool, key=lambda x: x["trip_work_minutes"], reverse=True)

    workdays = []

    for trip in trip_pool:
        best_day_idx = None
        best_remaining_time = None

        for i, wd in enumerate(workdays):
            current_minutes = sum(t["trip_work_minutes"] for t in wd)

            if current_minutes + trip["trip_work_minutes"] <= max_minutes:
                remaining_time = max_minutes - (current_minutes + trip["trip_work_minutes"])

                # Best fit: choose the day that leaves the least remaining time
                if best_remaining_time is None or remaining_time < best_remaining_time:
                    best_remaining_time = remaining_time
                    best_day_idx = i

        if best_day_idx is None:
            # Start a new workday
            workdays.append([trip])
        else:
            # Append to best existing day
            workdays[best_day_idx].append(trip)

    return workdays
```

```python
trips = days

workdays = pack_trips_into_workdays(
    trips,
    regular_customers,
    warehouse_lat,
    warehouse_lon,
    MAX_MINUTES,
    AVG_SPEED_MPH
)

for wd_idx, wd in enumerate(workdays, start=1):
    print(f"\nWorkday {wd_idx}")
    total_minutes = sum(t["trip_work_minutes"] for t in wd)
    print(f"  Total work minutes: {total_minutes:.2f}")
    for t_idx, trip in enumerate(wd, start=1):
        print(f"  Trip {t_idx}: {trip['customers']} | {trip['trip_work_minutes']:.2f} min")
```

```python
detailed_rows = []

for workday_num, workday in enumerate(workdays, start=1):
    current_day_time = 8 * 60  # 08:00 开始一天

    for trip_num, trip in enumerate(workday, start=1):
        trip_customer_ids = trip["customers"]

        trip_df = regular_customers[
            regular_customers["Customer ID"].isin(trip_customer_ids)
        ].copy()

        trip_df["route_order"] = trip_df["Customer ID"].map(
            {cust_id: i for i, cust_id in enumerate(trip_customer_ids)}
        )
        trip_df = trip_df.sort_values("route_order").reset_index(drop=True)

        # Preparation/Loading at the start of each trip: 23 minutes.
        current_day_time += 23

        prev_lat, prev_lon = warehouse_lat, warehouse_lon

        for stop_order, (_, row) in enumerate(trip_df.iterrows(), start=1):
            drive_miles = approx_miles(prev_lat, prev_lon, row["Lat"], row["Lon"])
            drive_minutes = drive_miles / AVG_SPEED_MPH * 60

            arrival_time = current_day_time + drive_minutes
            waiting_time = 0

            stop_fixed = 14
            service_minutes = row["Total_Service_Time"] + stop_fixed
            finish_time = arrival_time + service_minutes

            detailed_rows.append({
                "Workday": workday_num,
                "Trip_Number": trip_num,
                "Stop_Order": stop_order,
                "Customer ID": row["Customer ID"],
                "Customer Address": row["Customer Address"],
                "Arrival_Minutes": arrival_time,
                "Finish_Minutes": finish_time,
                "Arrival_Time": minutes_to_time(arrival_time),
                "Finish_Time": minutes_to_time(finish_time),
                "Drive_Miles_From_Previous": drive_miles,
                "Drive_Minutes_From_Previous": drive_minutes,
                "Waiting_Minutes": waiting_time,
                "Service_Minutes": service_minutes,
                "Total_Volume": row["Total_Volume"],
                "Total_Weight": row["Total_Weight"],
                "Lat": row["Lat"],
                "Lon": row["Lon"]
            })

            current_day_time = finish_time
            prev_lat, prev_lon = row["Lat"], row["Lon"]

        # Return to the warehouse after each trip.
        return_miles = approx_miles(prev_lat, prev_lon, warehouse_lat, warehouse_lon)
        return_minutes = return_miles / AVG_SPEED_MPH * 60
        return_arrival = current_day_time + return_minutes

        detailed_rows.append({
            "Workday": workday_num,
            "Trip_Number": trip_num,
            "Stop_Order": "Return",
            "Customer ID": "WAREHOUSE",
            "Customer Address": "590 20th St, San Francisco, CA 94107",
            "Arrival_Minutes": return_arrival,
            "Finish_Minutes": return_arrival,
            "Arrival_Time": minutes_to_time(return_arrival),
            "Finish_Time": minutes_to_time(return_arrival),
            "Drive_Miles_From_Previous": return_miles,
            "Drive_Minutes_From_Previous": return_minutes,
            "Waiting_Minutes": 0,
            "Service_Minutes": 0,
            "Total_Volume": 0,
            "Total_Weight": 0,
            "Lat": warehouse_lat,
            "Lon": warehouse_lon
        })

        current_day_time = return_arrival

schedule_detailed = pd.DataFrame(detailed_rows)
schedule_detailed["Stop_Sort"] = schedule_detailed.groupby(["Workday", "Trip_Number"]).cumcount() + 1

display(schedule_detailed.head(50))
```

```python
LABOR_PER_HOUR = 104
TRUCK_COST_PER_MILE = 1.205
CO2_PER_MILE = 1.69
CO2_IDLE_PER_HOUR = 3.97

daily_summary = schedule_detailed.groupby("Workday").agg(
    Total_Drive_Miles=("Drive_Miles_From_Previous", "sum"),
    Total_Drive_Minutes=("Drive_Minutes_From_Previous", "sum"),
    Total_Waiting_Minutes=("Waiting_Minutes", "sum"),
    Total_Service_Minutes=("Service_Minutes", "sum"),
    Stops=("Customer ID", lambda x: (x != "WAREHOUSE").sum()),
    Trips=("Trip_Number", "nunique")
).reset_index()

# Each trip includes a 23-minute preparation period.
trip_counts = schedule_detailed[schedule_detailed["Customer ID"] == "WAREHOUSE"].groupby("Workday").size().reset_index(name="Trip_Count")
daily_summary = daily_summary.merge(trip_counts, on="Workday", how="left")

daily_summary["Total_Work_Minutes"] = (
    daily_summary["Trip_Count"] * 23
    + daily_summary["Total_Drive_Minutes"]
    + daily_summary["Total_Service_Minutes"]
)

daily_summary["Labor_Cost"] = (daily_summary["Total_Work_Minutes"] / 60) * LABOR_PER_HOUR
daily_summary["Truck_Cost"] = daily_summary["Total_Drive_Miles"] * TRUCK_COST_PER_MILE
daily_summary["Total_Cost"] = daily_summary["Labor_Cost"] + daily_summary["Truck_Cost"]

daily_summary["CO2_Drive"] = daily_summary["Total_Drive_Miles"] * CO2_PER_MILE
daily_summary["CO2_Idle"] = (daily_summary["Total_Service_Minutes"] / 60) * CO2_IDLE_PER_HOUR
daily_summary["Total_CO2"] = daily_summary["CO2_Drive"] + daily_summary["CO2_Idle"]

daily_summary["Utilization_%"] = daily_summary["Total_Work_Minutes"] / 540

display(daily_summary)
```

## Naive Baseline Comparison (Optimization Impact)

To quantify the value of the optimization, we compare the optimized plan against a **naive no-consolidation baseline**: one dedicated round trip per order (no batching, no route sequencing), costed with the **same** labor, fuel, and CO2 model and identical service times. Only routing efficiency differs.

```python
PREP_MIN = 23

# --- Optimized totals (from daily_summary) ---
opt_miles = daily_summary['Total_Drive_Miles'].sum()
opt_cost  = daily_summary['Total_Cost'].sum()
opt_co2   = daily_summary['Total_CO2'].sum()
opt_trips = int(daily_summary['Trip_Count'].sum())
service_min_total = daily_summary['Total_Service_Minutes'].sum()
idle_co2_total = (service_min_total / 60) * CO2_IDLE_PER_HOUR

# --- Naive baseline: one dedicated round trip per regular customer ---
naive_miles = regular_customers['Miles_From_Warehouse'].sum() * 2
naive_trips = len(regular_customers)
naive_drive_min = naive_miles * 2   # 30 mph (minutes = miles * 2)
naive_work_min  = naive_trips * PREP_MIN + naive_drive_min + service_min_total
naive_labor = naive_work_min / 60 * LABOR_PER_HOUR
naive_truck = naive_miles * TRUCK_COST_PER_MILE
naive_cost  = naive_labor + naive_truck
naive_co2   = naive_miles * CO2_PER_MILE + idle_co2_total

comparison = pd.DataFrame({
    'Scenario': ['Naive (no consolidation)', 'Optimized'],
    'Trips': [naive_trips, opt_trips],
    'Drive_Miles': [naive_miles, opt_miles],
    'Operating_Cost': [naive_cost, opt_cost],
    'Total_CO2_kg': [naive_co2, opt_co2],
})
for col in ['Drive_Miles', 'Operating_Cost', 'Total_CO2_kg']:
    saved = (comparison[col][0] - comparison[col][1]) / comparison[col][0] * 100
    print(f'{col:16} reduced {saved:5.1f}%  ({comparison[col][0]:.1f} -> {comparison[col][1]:.1f})')
print(f'Annualized cost saved (x52): ${(naive_cost-opt_cost)*52:,.0f}')
print(f'Annualized CO2 saved  (x52): {(naive_co2-opt_co2)*52:,.0f} kg')

comparison.to_csv('baseline_comparison_final.csv', index=False)
display(comparison)
```
